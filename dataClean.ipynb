```python
df = pd.read_csv("text_segments.csv", encoding='unicode_escape');
```

```python
grouped_df = df.groupby(["doc_name","pagenum"]).apply(lambda x : x.sort_values('pagenum')).reset_index(drop=True)
```

```python
merged_rows = grouped_df.groupby(["doc_name","pagenum"]).agg({
    "text": lambda x : "".join(map(str,x))
}).reset_index()
```

```python
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
pd.set_option("display.max_colwidth",None)

```

```python
text_array = []
```

```python
for index, row in merged_rows.iterrows():
    doc_name = row["doc_name"]
    page_num = row["pagenum"]
    content = row["text"]
    
    text_array.append({
        "doc_name" : doc_name,
        "page_num" : page_num,
        "content" : content
    })
```

```python
text_array
```

```python
file1 = []
file2 = []
file3 = []

for i in range(len(text_array)):
    doc_name = text_array[i]["doc_name"]
    if doc_name == "dairy-farm_ar2020pdf":
        file1.append({
            "Title" : text_array[i]["doc_name"],
            "PageNum" : text_array[i]["page_num"],
            "content" : text_array[i]["content"] 
        })
    elif doc_name == "with-highlights-comments-barclays-country-snapshot-2021pdf":
        file2.append({
            "Title" : text_array[i]["doc_name"],
            "PageNum" : text_array[i]["page_num"],
            "content" : text_array[i]["content"] 
        })
    else:
        file3.append({
            "Title" : text_array[i]["doc_name"],
            "PageNum" : text_array[i]["page_num"],
            "content" : text_array[i]["content"] 
        })
```

```python
current_doc_name = ""
current_pagenum = 0

for item in file3:
    if item["Title"] != current_doc_name or item["PageNum"] != current_pagenum:
        pdf.add_page()
        current_doc_name = item["Title"]
        current_pagenum = item["PageNum"]
    
    pdf.add_font("Arial","",r"C:\Users\Hp\Downloads\arial\arial.ttf",uni=True)    
    pdf.set_font("Arial",size = 12)
    pdf.multi_cell(200,10,txt=item["content"].encode('latin-1', 'ignore').decode('utf'))
    
pdf.output("file3.pdf")
```
